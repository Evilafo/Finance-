```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Les données historiques de l'action NVIDIA (NVDA)
nvda = yf.Ticker("NVDA")
historical_data = nvda.history(period="1y")  # On ajuste la période sur 1 an

historical_data = historical_data.rename(columns={
    'Open': 'Ouverture',
    'High': 'Haut',
    'Low': 'Bas',
    'Close': 'Fermeture',
    'Volume': 'Volume',
    'Dividends': 'Dividendes',
    'Stock Splits': 'Actions fractionnées'
})

historical_data.head()
```

```python
# L'évolution du prix de l'action
plt.figure(figsize=(10, 6))
plt.plot(historical_data.index, historical_data['Fermeture'], label='Prix de l\'action NVDA')
plt.axvline(pd.Timestamp('2025-01-27'), color='red', linestyle='dotted', label='Annonce de DeepSeek')

# Annotation ajustée
plt.annotate('Annonce DeepSeek : 2025-01-27',
             xy=(pd.to_datetime('2025-01-27'), 115),
             xytext=(pd.to_datetime('2024-10-20'), 100),
             arrowprops=dict(facecolor='yellow', shrink=0.01))

plt.title('Évolution du prix de l\'action NVIDIA sur 1 an')
plt.xlabel('Date')
plt.ylabel('Prix de l\'action (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# L'évolution du prix de l'action
plt.figure(figsize=(10, 6))
plt.plot(historical_data.index, historical_data['Fermeture'], label='Prix de l\'action NVDA')
plt.axvline(pd.Timestamp('2025-01-27'), color='red', linestyle='dotted', label='Annonce de DeepSeek')
plt.title('Évolution du prix de l\'action NVIDIA sur 1 an')
plt.xlabel('Date')
plt.ylabel('Prix de l\'action (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Les données historiques de AMD, Intel, et Qualcomm
amd = yf.Ticker("AMD")
intel = yf.Ticker("INTC")
qualcomm = yf.Ticker("QCOM")

# Les données de 1 an
amd_data = amd.history(period="1y")
intel_data = intel.history(period="1y")
qualcomm_data = qualcomm.history(period="1y")

# Les courbes comparatives
plt.figure(figsize=(10, 6))
plt.plot(historical_data.index, historical_data['Fermeture'], label='NVIDIA')
plt.plot(amd_data.index, amd_data['Close'], label='AMD')
plt.plot(intel_data.index, intel_data['Close'], label='Intel')
plt.plot(qualcomm_data.index, qualcomm_data['Close'], label='Qualcomm')
plt.axvline(pd.Timestamp('2025-01-27'), color='red', linestyle='dotted', label='Annonce de DeepSeek')
plt.title('Comparaison des actions : NVIDIA, AMD, Intel, Qualcomm')
plt.xlabel('Date')
plt.ylabel('Prix de l\'action (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calcul de la moyenne mobile sur 50 jours et 200 jours
historical_data['MA50'] = historical_data['Fermeture'].rolling(window=50).mean()
historical_data['MA200'] = historical_data['Fermeture'].rolling(window=200).mean()

# Le prix de l'action avec les moyennes mobiles
plt.figure(figsize=(10, 6))
plt.plot(historical_data.index, historical_data['Fermeture'], label='Prix de l\'action NVDA')
plt.plot(historical_data.index, historical_data['MA50'], label='Moyenne mobile 50 jours')
plt.plot(historical_data.index, historical_data['MA200'], label='Moyenne mobile 200 jours')
plt.axvline(pd.Timestamp('2025-01-27'), color='red', linestyle='dotted', label='Annonce de DeepSeek')
plt.title('Prix de l\'action NVIDIA avec moyennes mobiles')
plt.xlabel('Date')
plt.ylabel('Prix de l\'action (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
